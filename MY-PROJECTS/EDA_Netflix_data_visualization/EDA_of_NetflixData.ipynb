```python
import pandas as pd
```

```python
netflix_df = pd.read_csv('/content/drive/MyDrive/netflix_titles.csv')
```

```python
display(netflix_df.head())
```

```python
print("1. Displaying concise summary of the DataFrame:\n")
netflix_df.info()

print("\n2. Counting and displaying missing values for each column:\n")
print(netflix_df.isnull().sum())

print("\n3. Filling missing values in 'director', 'cast', and 'country' with 'Unknown'...")
netflix_df['director'].fillna('Unknown', inplace=True)
netflix_df['cast'].fillna('Unknown', inplace=True)
netflix_df['country'].fillna('Unknown', inplace=True)
print("Missing values in 'director', 'cast', and 'country' filled.")

print("\n4. Converting 'date_added' to datetime objects and extracting 'year_added'...")
netflix_df['date_added'] = pd.to_datetime(netflix_df['date_added'], errors='coerce')
netflix_df['year_added'] = netflix_df['date_added'].dt.year

netflix_df['year_added'].fillna(0, inplace=True)
netflix_df['year_added'] = netflix_df['year_added'].astype(int)
print("'date_added' converted to datetime and 'year_added' extracted and filled.")

print("\nUpdated missing values after cleaning:\n")
print(netflix_df.isnull().sum())

print("\nFirst 5 rows of the DataFrame after cleaning:\n")
display(netflix_df.head())
```

```python
import matplotlib.pyplot as plt

# Calculate the value counts for the 'type' column
content_type_distribution = netflix_df['type'].value_counts()

# Create a pie chart
plt.figure(figsize=(8, 8))
plt.pie(content_type_distribution, labels=content_type_distribution.index, autopct='%1.1f%%', startangle=90, colors=['skyblue', 'lightcoral'])
plt.title('Distribution of Content Types on Netflix')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()
```

```python
import matplotlib.pyplot as plt

# 1. Calculate the number of releases for each 'release_year'
releases_per_year = netflix_df['release_year'].value_counts().sort_index()

# 2. Create a line plot to visualize the trend of content releases over the years
plt.figure(figsize=(14, 7))
releases_per_year.plot(kind='line', marker='o', color='purple')

# 3. Set the x-axis label as 'Release Year' and the y-axis label as 'Number of Releases'
plt.xlabel('Release Year')
plt.ylabel('Number of Releases')

# 4. Add a title to the plot
plt.title('Netflix Content Releases Over Time')

# Improve readability of x-axis ticks
plt.xticks(rotation=45)
plt.grid(True)

# 5. Display the plot
plt.show()
```
